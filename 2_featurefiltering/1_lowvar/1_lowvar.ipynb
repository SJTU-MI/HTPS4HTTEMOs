```python
# Introduction of data
# Excluding low variance data
import pandas as pd
file_path = r'data4PF.csv'
all_data = pd.read_csv(file_path)
```

```python
all_data
```

```python
X = all_data.iloc[:,3:318]
y = all_data.iloc[:,0:3]
```

```python
y
```

```python
X
```

```python
import numpy as np
X_var = pd.DataFrame(X.var())
file_path = r'PF_data_var.csv'
X_var.to_csv(file_path, index = True)
```

```python
from sklearn.feature_selection import VarianceThreshold
vt = VarianceThreshold(threshold = 0.01)
X_selected = vt.fit_transform(X)
lowvariance_data = pd.DataFrame(X_selected)
X_selected.shape
```

```python
all_name = X.columns.values.tolist() 
select_name_index0 = vt.get_support(indices=True) 
select_name0 = []
for i in select_name_index0:
    select_name0.append(all_name[i])
```

```python
len(select_name0)
```

```python
lowvariance_data.columns = select_name0
file_path = r'PF_data_lowvar.csv'
lowvariance_data_y = pd.concat((y,lowvariance_data),axis = 1)
lowvariance_data_y.to_csv(file_path, index = False)
```
