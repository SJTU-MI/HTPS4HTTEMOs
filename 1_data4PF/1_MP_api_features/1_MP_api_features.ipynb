```python
# import package

from pymatgen.ext.matproj import MPRester
import pandas as pd
```

```python
# Import material id

def ReadTxtName(rootdir):
    lines = []
    with open(rootdir, 'r') as file_to_read:
        while True:
            line = file_to_read.readline()
            if not line:
                break
            line = line.strip('\n')
            lines.append(line)
    return lines

resultpath='PF_mpid.txt'
lineslist=ReadTxtName(resultpath)
```

```python
# API from MP 
# Please use your own key

mpr = MPRester("Please use your own API key of MP")
mp_decode = False

outputdata = mpr.query(criteria={'task_id':{'$in':lineslist}}, 
                       properties=[
                           'task_id',
                           'pretty_formula',
                           'spacegroup.number',
                           'structure',
                           'nsites',
                           'volume',
                           'band_gap',
                           'density',
                           'e_above_hull',
                           'efermi',
                           'energy',
                           'final_energy_per_atom',
                           'formation_energy_per_atom',
                           'nelements'
                       ])
```

```python
# feature processing


download_data = pd.DataFrame(outputdata)
API_descriptor = download_data.iloc[:,0:3]
a = []
b = []
c = []
alpha = []
beta = []
gamma = []
for list_mat in outputdata:
    a.append(list_mat['structure'].lattice.abc[0])
    b.append(list_mat['structure'].lattice.abc[1])
    c.append(list_mat['structure'].lattice.abc[2])
    alpha.append(list_mat['structure'].lattice.angles[0])
    beta.append(list_mat['structure'].lattice.angles[1])
    gamma.append(list_mat['structure'].lattice.angles[2])

API_descriptor['a'] = pd.DataFrame(a)
API_descriptor['b'] = pd.DataFrame(b)
API_descriptor['c'] = pd.DataFrame(c)
API_descriptor['alpha'] = pd.DataFrame(alpha)
API_descriptor['beta'] = pd.DataFrame(beta)
API_descriptor['gamma'] = pd.DataFrame(gamma)
API_descriptor['volume'] = download_data['volume'] 
API_descriptor['volume/atom'] = download_data['volume']/download_data['nsites']
API_descriptor['band_gap'] = download_data['band_gap']
API_descriptor['density'] = download_data['density']
API_descriptor['e_above_hull'] = download_data['e_above_hull']
API_descriptor['e_above_hull/atom'] = download_data['e_above_hull']/download_data['nsites']
API_descriptor['efermi'] = download_data['efermi']
API_descriptor['energy'] = download_data['energy']
API_descriptor['energy/atom'] = download_data['energy']/download_data['nsites']
API_descriptor['final_energy_per_atom'] = download_data['final_energy_per_atom']
API_descriptor['formation_energy_per_atom'] = download_data['formation_energy_per_atom']
API_descriptor['nelements'] = download_data['nelements']
API_descriptor['nsites'] = download_data['nsites']
API_descriptor.columns = ['ID','Name','spacegroup','a','b','c','alpha','beta','gamma',
                          'volume','volume/atom','band_gap','density','e_above_hull',
                          'e_above_hull/atom','efermi','energy','energy/atom','final_energy_per_atom',
                          'formation_energy_per_atom','nelements','nsites']
API_descriptor = API_descriptor.set_index(['ID'])
```
