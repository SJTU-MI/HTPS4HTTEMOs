```python
import json
import pandas as pd
filepath = 'melt_data.csv'

with open('melt_result.json', encoding='utf-8') as f:
    line = f.readline()
    d = json.loads(line)
    f.close()
```

```python
elem_list = []
comp_list = []
melt_list = []
for material in d:
    elem_list.append(material['elments'])
    comp_list.append(material['compositions'])
    melt_list.append(material['melting temperature'])
    print(material['elments'],material['compositions'],material['melting temperature'])
```

```python
pd_elem_list = pd.DataFrame(elem_list)
pd_comp_list = pd.DataFrame(comp_list)
pd_melt_list = pd.DataFrame(melt_list)

melt_data = pd.concat([pd_elem_list,pd_comp_list,pd_melt_list],axis=1)
melt_data.columns = ['elem1','elem2','elem3','elem4','comp1','comp2','comp3','comp4','melt_point']
melt_data.to_csv('melt_data.csv')
```
