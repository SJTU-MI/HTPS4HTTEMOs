```python
# Please use your own API key

from pymatgen.ext.matproj import MPRester
mpr = MPRester("Please use your own API key")
mp_decode = False
```

```python
import pandas as pd
def ReadTxtName(rootdir):
    lines = []
    with open(rootdir, 'r') as file_to_read:
        while True:
            line = file_to_read.readline()
            if not line:
                break
            line = line.strip('\n')
            lines.append(line)
    return lines

resultpath='mpid4melting_point_pre.txt'
lineslist=ReadTxtName(resultpath)
```

```python
outputdata = mpr.query(criteria={'task_id':{'$in':lineslist}}, 
                       properties=[
                           'task_id',
                           'pretty_formula',
                           'nelements',
                           'unit_cell_formula'
                       ])
```

```python
import json
data_json_list = []
data_task_id_list = []
data_pretty_formula_list = []
for list_mat in outputdata:
    data_task_id_list.append(list_mat['task_id'])
    data_pretty_formula_list.append(list_mat['pretty_formula'])    
    element_num = len(list_mat['unit_cell_formula'])
    elem_list = []
    comp_list = []
    for composition in list_mat['unit_cell_formula']:
        elem_list.append(composition)
        comp_list.append(list_mat['unit_cell_formula'][composition])
    print(list_mat['task_id'],list_mat['pretty_formula'],elem_list,comp_list,element_num)
    if element_num == 1:
        data = {"elem1":elem_list[0],"comp1":comp_list[0],
                "elem2":"","comp2":"",
                "elem3":"","comp3":"",
                "elem4":"","comp4":""}
    elif element_num == 2:
        data = {"elem1":elem_list[0],"comp1":comp_list[0],
                "elem2":elem_list[1],"comp2":comp_list[1],
                "elem3":"","comp3":"",
                "elem4":"","comp4":""}
    elif element_num == 3:
        data = {"elem1":elem_list[0],"comp1":comp_list[0],
                "elem2":elem_list[1],"comp2":comp_list[1],
                "elem3":elem_list[2],"comp3":comp_list[2],
                "elem4":"","comp4":""}   
    elif element_num == 4:
        data = {"elem1":elem_list[0],"comp1":comp_list[0],
                "elem2":elem_list[1],"comp2":comp_list[1],
                "elem3":elem_list[2],"comp3":comp_list[2],
                "elem4":elem_list[3],"comp4":comp_list[3]}
    with open("melt_data.json", 'w') as file_obj:
        data_json_list.append(data)
        json.dump(data_json_list, file_obj)
```
